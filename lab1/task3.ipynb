## User stored in dictionary and based crud type ops 


```python
# dic to save user
user_data = {}

# add func 
def add_user():
    name = input("Name: ")
    contact = input("Contact number: ")
    email = input("Email address: ")
    user_data[name] = {'contact': contact, 'email': email}
    print(f"\n{name} added successfully!\n")

# search func 
def search_user():
    name = input("Search: ")
    if name in user_data:
        print(f"\nFound: {name}")
        print("Contact:", user_data[name]['contact'])
        print("Email:", user_data[name]['email'], "\n")
    else:
        print(f"\n{name} not found in records.\n")

# delete func
def delete_user():
    name = input("Delete: ")
    if name in user_data:
        del user_data[name]
        print(f"\n{name} deleted successfully!\n")
    else:
        print(f"\n{name} not found in records.\n")

# show all users func 
def show_all_users():
    if not user_data:
        print("\nNo records found.\n")
    else:
        print("\n--- All Users ---")
        for name, info in user_data.items():
            print(f"Name: {name}, Contact: {info['contact']}, Email: {info['email']}")
        print()

# program
while True:
    print("1. Add User")
    print("2. Search User")
    print("3. Delete User")
    print("4. Show All Users")
    print("5. Exit")
    choice = input("Enter your choice (1-5): ")

    if choice == '1':
        add_user()
    elif choice == '2':
        search_user()
    elif choice == '3':
        delete_user()
    elif choice == '4':
        show_all_users()
    elif choice == '5':
        print("Exiting program...")
        break
    else:
        print("Invalid choice. Please try again.\n")
```
